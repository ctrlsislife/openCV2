```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/lena256.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg1.shape)
cv2_imshow(gimg1)
#plt.imshow(gimg1, cmap='gray')
```

```python
# 평균 필터 만들기
kernal3 = np.ones((3,3), np.float32)/9       # 3x3 의 111111111을 각각 9로 나눠 1/9값 가지게함
```

```python
outimg = cv2.filter2D(gimg1, -1, kernal3)
cv2_imshow(outimg)
```

```python
# 평균 필터 만들기
kernal5 = np.ones((5,5), np.float32)/25       # 5x5배열 의 111111111...을 각각 25로 나눠 1/25값 가지게함
```

```python
outimg = cv2.filter2D(gimg1, -1, kernal5)
cv2_imshow(outimg)
```

```python
outimg33 = cv2.filter2D(outimg, -1, kernal3)  # 평균 필터를 여러번 적용하면 더 블러 처리 되게 된다.
cv2_imshow(outimg33)
```

```python
kernal3[0][0] = 1/9  # 직접 커널 구현해보기 ( Low Frequency Filter)
kernal3[0][1] = 1/9
kernal3[0][2] = 1/9
kernal3[1][0] = 1/9
kernal3[1][1] = 1/9
kernal3[1][2] = 1/9
kernal3[2][0] = 1/9
kernal3[2][1] = 1/9
kernal3[2][2] = 1/9
```

```python
outimg = cv2.filter2D(gimg1, -1, kernal3)  # 직접 커널을 구현해도 kernal3 라이브러리를 사용한 것과 동일한 것을 확인
cv2_imshow(outimg)
```
