 RGB 컬러 영상을 Red, Green, Blue 3 채널로 분리하기

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/lena256.jpg')
```

```python
# BGR 채널 순서를 RGB 순서로 변경
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# RGB_img를 Red Green Blue 채널로 분할하기
RED_img, Green_img, Blue_img = cv2.split(RGB_img)
```

```python
# 원본 그림을 화면에 출력하기
plt.title("Original Image")
plt.imshow(RGB_img)
```

```python
# R채널만 출력하기
RGB_img[:, :, 0] = RED_img
RGB_img[:, :, 1] = RED_img
RGB_img[:, :, 2] = RED_img
plt.title("Red Channel")
plt.imshow(RGB_img)
plt.title("Red Channel")
plt.imshow(RGB_img)
```

```python
# G 채널만 출력하기
RGB_img[:, :, 0] = Green_img
RGB_img[:, :, 1] = Green_img
RGB_img[:, :, 2] = Green_img
plt.title("Green Channel")
plt.imshow(RGB_img)
```

```python
# B 채널만 출력하기
RGB_img[:, :, 0] = Blue_img
RGB_img[:, :, 1] = Blue_img
RGB_img[:, :, 2] = Blue_img
plt.title("Blue Channel")
plt.imshow(RGB_img)
```
