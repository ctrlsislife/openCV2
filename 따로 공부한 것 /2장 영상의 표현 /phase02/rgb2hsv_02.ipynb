RGB 컬러 영상을 H(색상) S(채도) V(명도) 채널로 분리하기

```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/lena256.jpg')
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR -> RGB
```

```python
HSV_img = cv2.cvtColor(RGB_img, cv2.COLOR_BGR2HSV)   #RGB -> HSV
```

```python
H_img, S_img, V_img = cv2.split(HSV_img) # h , s , v 채널로 나누기
```

```python
plt.figure( figsize=(20,20) ) # 영상 크기 키우기
plt.subplot(1, 4, 1) # 1행 4열에서 1번째 열
plt.title("Original Image")
plt.imshow(RGB_img)
plt.axis("off")  # 그리드 삭제

plt.subplot(1, 4, 2) # 1행 4열에서 2번째 열
RGB_img[:,:,0] = H_img
RGB_img[:,:,1] = H_img
RGB_img[:,:,2] = H_img
plt.title("H Channel")
plt.imshow(RGB_img)
plt.axis("off")

plt.subplot(1, 4, 3) # 1행 4열에서 3번째
RGB_img[:,:,0] = S_img
RGB_img[:,:,1] = S_img
RGB_img[:,:,2] = S_img
plt.title("S Channel")
plt.imshow(RGB_img)
plt.axis("off")

plt.subplot(1, 4, 4) # 1행 4열에서 4번째
RGB_img[:,:,0] = V_img
RGB_img[:,:,1] = V_img
RGB_img[:,:,2] = V_img
plt.title("V Channel")
plt.imshow(RGB_img)
plt.axis("off")

```
