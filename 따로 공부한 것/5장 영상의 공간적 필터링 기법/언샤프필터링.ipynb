```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/Lena-Gaussian-noise1.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # 컬러 영상을 그레이 영상으로
```

```python
# 5x5 커널 적용
mean_img = cv2.blur(gray_img, (5, 5))
edge_img = cv2.addWeighted(gray_img, 1.0, mean_img, -1.0, 0)
output_img =cv2.addWeighted(gray_img, 1.0, edge_img,3.0,0)
```

```python
# 출력
plt.figure(figsize=(20,20))
plt.subplot(1, 3, 1)
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1, 3, 2)
plt.title("Edge Image")
plt.imshow(edge_img, cmap='gray')
plt.axis("off")
plt.subplot(1, 3, 3)
plt.title("Unsharped Image")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
