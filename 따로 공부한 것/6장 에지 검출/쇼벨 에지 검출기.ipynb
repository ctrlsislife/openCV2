```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/lena.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # 컬러 영상을 그레이 영상으로
```

```python
# sobel operator
x = cv2.Sobel(gray_img, -1, 1, 0)
y = cv2.Sobel(gray_img, -1, 0, 1)
```

```python
# uint8 (0~255) 타입으로 변경하고 영상 병합
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)
```

```python
# 그림을 화면에 출력
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Sobel Edge Detection")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
