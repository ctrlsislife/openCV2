```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/airplane.bmp') # 영상 불러오기
```

```python
height,width = img.shape[:2]
# 회전점을 영상 모서리에서 영상 중심으로 변경
M = cv2.getRotationMatrix2D(((width-1)/2.0,(height-1)/2.0),90,1)
output_img = cv2.warpAffine(img,M,(width*1,height*1),flags = cv2.INTER_LINEAR)
```

```python
# 출력
print("Original Image",img.shape[:2])
cv2_imshow(img)
print("Rotation Transformation",output_img.shape[:2])
cv2_imshow(output_img)
```
