```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/airplane.bmp') # 영상 불러오기
```

```python

# BGR 2 RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height,width = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]]) # 가로 100 픽셀, 세로 50 픽셀 만큼 영상 이동
out1 = cv2.warpAffine(img,M,(width,height)) # 변환행렬을 잉요한 영상 이동 결과 할당
M = np.float32([[1,0,-50],[0,1,-50]])  # 가로 -50 픽셀, 세로 -50 픽셀 만큼 영상 이동
out2 = cv2.warpAffine(img,M,(width,height)) # 변환행렬을 이용한 영상 이동 결과 할당
```

```python
# 출력
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")
plt.subplot(1,3,2)
plt.title("Translation Image (x direction : 100 , y direction : 50)")
plt.imshow(out1)
plt.axis("off")
plt.subplot(1,3,3)
plt.title("Translation Image (x direction : -50 , y direction : -50)")
plt.imshow(out2)
plt.axis("off")
plt.show()
```
