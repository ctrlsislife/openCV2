```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/airplane.bmp') # 영상 불러오기
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR을 RGB로 변환
```

```python
# 수직 반전(vertical)
flipVertical=cv2.flip(img, 0)

# 수평 반전(Horizontal)
flipHorizontal=cv2.flip(img, 1)

# 상하좌우 반전(Flip Both)
flipBoth=cv2.flip(img, -1)
```

```python
# 출력
plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Flip Vertical Image")
plt.imshow(flipVertical)
plt.axis("off")
plt.subplot(1,4,3)
plt.title("Flip Horizontal Image")
plt.imshow(flipHorizontal)
plt.axis("off")
plt.subplot(1,4,4)
plt.title("Flip Both Image")
plt.imshow(flipBoth)
plt.axis("off")
plt.show()
```
