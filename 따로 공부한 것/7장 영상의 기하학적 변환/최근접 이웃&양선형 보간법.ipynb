```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/flower.bmp')
# 최근접 이웃 보간법 구현
output_img = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_NEAREST)

# 양선형 보간법 구현
output_img2 = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_LINEAR)
```

```python
# 출력
print("Original Image",img.shape[:2])
cv2_imshow(img)

print("Nearest Neighbor Interpolation", output_img.shape[:2])
cv2_imshow(output_img)

print("Bilinear Interpolation", output_img2.shape[:2])
cv2_imshow(output_img2)
```
