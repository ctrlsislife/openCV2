```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/airplane.bmp') # 영상 불러오기
```

```python
# 영상 크기변환
resize = cv2.resize(img,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_LINEAR)
height,width = resize.shape[:2]
```

```python
# 회전점을 영상 모서리에서 영상의 중심으로 변경
M = cv2.getRotationMatrix2D(((width-1)/2.0,(height-1)/2.0),-15,1)
rotate = cv2.warpAffine(resize,M,(width,height))
```

```python
# 영상이동변환
M = np.float32([[1,0,30],[0,1,+20]])
translate = cv2.warpAffine(rotate,M,(width,height))
```

```python
# 출력
print("Original Image")
cv2_imshow(img)

print("Scaling Image")
cv2_imshow(resize)

print("Rotation (45 degree) image")
cv2_imshow(rotate)

print("Translation Image(x :+30 , y : +20)")
cv2_imshow(translate)
```
