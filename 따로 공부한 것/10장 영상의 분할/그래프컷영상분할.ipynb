```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/flower.bmp') # 영상 불러오기
```

```python
img =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img)

mask = np.zeros(img.shape[:2], np.uint8)
# 영상 크기를 받아오고 영상 크기 만큼의 리스트를 모두 0으로 초기화한 마스크

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
# 배경 및 전경 모델 , 크기가 (1,65)인 두 개의 np.float64 유형 배열을 만듦

rec = (50,20,210,180)
# 전경과 배경을 나눌 사각형 영역 설정

cv2.grabCut
(img,mask,rec,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
# 그랩컷 반복회수가 5회

mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
# 마스크 수정

imgout = img*mask2[:,:,np.newaxis]
# 사각형과 그랩컷 영상 출력



```

```python
# 출력
cv2.rectangle(img, (50,20), (260,200), (255,0,0))
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.title('Input')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('GrabCut')
plt.imshow(imgout)
plt.show()
```
