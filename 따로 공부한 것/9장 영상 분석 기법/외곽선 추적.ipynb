```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python

img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/milkdrop.bmp') # 영상 불러오기
```

```python
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# threshold 이용해 이진화 이미지로 변경
ret, thresh = cv2.threshold(gray_img, 127, 255, 0)
# 그레이 영상에 대해 임계값 127을 설정하고 임계값보다 크면 픽셀값을 255로 변경, 그렇지 않다면 0으로 변경한다.

# contours는 point의 리스트 형태
# hierachy는 contours line의 계층 구조임
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
# 이진화 영상에 외곽선 점들을 추출하고 컨투어 포인트 리스트에 입력
```

```python
# 영상에 외곽선 그리기
image = cv2.drawContours(img,contours, -1, (0,0,255),1) # -1 : 전체 외곽선 그리기 , 1: 외곽선 굵기
print("Segmented image")
cv2_imshow(thresh)
print("Contour image")
cv2_imshow(image)
```
