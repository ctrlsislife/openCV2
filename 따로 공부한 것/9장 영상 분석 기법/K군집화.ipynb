```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randint(25,50,(25,2)) # 25~49 사이의 정수를 반환 , 25행 2열짜리 배열로
Y = np.random.randint(60,85,(25,2))


Z = np.vstack((X,Y)) # 두 배열 연결
Z = np.float32(Z) # 배열의 값을 실수로 변경
```

```python
# 기준 정의 후 kmeans() 함수에 적용
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret, label, center = cv2.kmeans(Z, 3, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
# 지정된 정확도에 도달하면 알고리즘 반복 중지, 최대 반복 횟수 10


A = Z[label.ravel() == 0] # 군집을 나눔, 군집 레이블이 0인 데이터만 A 배열에 넣기
B = Z[label.ravel() == 1]
```

```python
# 군집을 화면에 표시
plt.scatter(A[:,0], A[:,1], c = 'b') # 배열 A의 첫 번째 두번째 열값을 도표에 블루로 표시
plt.scatter(B[:,0], B[:,1], c = 'r')
plt.scatter(center[:,0], center[:,1], s = 200, c = 'y', marker = '*')
# 군집 중심값을 마커 *로 표시, 마커의 크기는 200 , 색상은 노랑
plt.xlabel('Height'), plt.ylabel('Weight')
plt.show()
```
