```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/char.bmp') # 영상 불러오기
```

```python
kernal = np.ones((3,3),np.uint8) # 1을 인수로 가지는 3x3 배열 생성
erosion1 = cv2.erode(img,kernal,iterations = 1)
```

```python
print("Original Image")
cv2_imshow(img)

print("Erosion 1 Image")
cv2_imshow(erosion1)
```
