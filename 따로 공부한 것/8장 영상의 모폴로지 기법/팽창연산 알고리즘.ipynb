```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/rice.bmp',0) # 영상 불러오기 0은 그레이스케일로 불러오기를 뜻함
```

```python
gray = img
height, width = gray.shape
```

```python
out = np.zeros((height+2, width+2),dtype=np.uint8)
out[1:height+1, 1:width+1] = gray.copy()
dilation = np.zeros_like(gray)

for i in range(height):
  for j in range(width):
    temp = out[i:i+3, j:j+3]
    dilation[i,j] = np.max(temp)

```

```python
cv2_imshow(gray)
cv2_imshow(dilation)
```
