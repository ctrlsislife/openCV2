```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/Picture4.png') # 영상 불러오기
img2 = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/rice.bmp') # 영상 불러오기
```

```python
kernal = np.ones((5,5),np.uint8)
```

```python
opening1 = cv2.morphologyEx(img1, cv2.MORPH_OPEN, kernal)
opening2 = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernal)
```

```python
print("Original image")
cv2_imshow(img1)
print("Opening image")
cv2_imshow(opening1)
print("Original Image")
cv2_imshow(img2)
print("Opening Image")
cv2_imshow(opening2)
```
