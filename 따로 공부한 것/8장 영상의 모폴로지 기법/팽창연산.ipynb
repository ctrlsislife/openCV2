```python

# Perform dilation
kernal = np.ones((3,3),np.uint8) # 1을 인수로 가진 3x3 배열 생성 , np.uint8 :부호 없는 정수형(unsigned integers)
dilation = cv2.dilate(img,kernal,iterations=1) #(팽창시킼ㄹ 이미지, 커널 크기, 반복 횟수)
```

```python

# Display the images
print("Original Image")
cv2_imshow(img)
print("Dilated Image")
cv2_imshow(dilation)
print("Edge image")
cv2_imshow(dilation-img)
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/char.bmp') # 영상 불러오기

# Perform dilation
kernal = np.ones((3,3),np.uint8) # 1을 인수로 가진 3x3 배열 생성 , np.uint8 :부호 없는 정수형(unsigned integers)
dilation = cv2.dilate(img,kernal,iterations=1) #(팽창시킼ㄹ 이미지, 커널 크기, 반복 횟수)

# Display the images
print("Original Image")
cv2_imshow(img)
print("Dilated Image")
cv2_imshow(dilation)
print("Edge image")
cv2_imshow(dilation-img)
```
