```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/aero2.bmp') # 영상 불러오기
```

```python
kernal = np.ones((5,5),np.uint8) # 1을 인수로 가지는 5x5 배열 생성
```

```python
# 모폴로지 닫힘 연산 수행
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernal)
```

```python
# 출력
print("Original Image")
cv2_imshow(img)
print("Closing Image")
cv2_imshow(closing)
```
