```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/shrinking.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg1.shape)
cv2_imshow(gimg1)
#plt.imshow(gimg1, cmap='gray')
```

```python
outImg = cv2.resize(gimg1, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST) # 인접한 것으로 반올림

cv2_imshow(outImg)
```

```python
outImg = cv2.resize(gimg1, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_LINEAR) #

cv2_imshow(outImg)
```

```python
outImg = cv2.resize(gimg1, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_NEAREST)

cv2_imshow(outImg)
```

```python
outImg = cv2.resize(gimg1, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_CUBIC) # 축소시 엣지가 사라짐 -> 인터폴레이션의 방법적 문제가 아님

cv2_imshow(outImg)
```

```python
blurImg = cv2.GaussianBlur(gimg1, (3,3), 1)
outImg = cv2.resize(blurImg, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_NEAREST)

cv2_imshow(outImg)
cv2_imshow(blurImg)
```
