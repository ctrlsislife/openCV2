```python
from google.colab import drive
drive.mount('/content/gdrive')
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/abcdef.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg1 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg1.shape)
cv2_imshow(gimg1)
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/a.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg2.shape)
cv2_imshow(gimg2)
```

```python
gBit = np.zeros((gimg1.shape[0], gimg1.shape[1]), dtype = np.ubyte) # 저장할 공간 할당
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    bitVal = gimg1[h,w] & gimg2[h,w]
    gBit[h,w] = bitVal

cv2_imshow(gBit)   # abcde 와 a를 and 연산시 a만 나오는 것 확인
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    bitVal = gimg1[h,w] | gimg2[h,w]
    gBit[h,w] = bitVal

cv2_imshow(gBit)   # abcde 와 a를 or 연산시 abcdef 나오는 것 확인
```

```python
img = cv2.imread('/content/gdrive/My Drive/ColabIP/Images/circle.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg2 = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg2.shape)
cv2_imshow(gimg2)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    bitVal = gimg1[h,w] & gimg2[h,w]
    gBit[h,w] = bitVal

cv2_imshow(gBit)  # lena 와 circle의 AND 연산
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    bitVal = gimg1[h,w] | gimg2[h,w]
    gBit[h,w] = bitVal

cv2_imshow(gBit)  # lena 와 circle의 OR 연산
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    bitVal = gimg1[h,w] ^ gimg2[h,w]
    gBit[h,w] = bitVal

cv2_imshow(gBit)  # lena 와 circle의 Exclusive OR 연산
```
